```python
import pandas as pd
```

# Raw dataset

```python
df = pd.read_excel(r"C:\Extras\Programs\My Projects\customer call data cleaning\dataset\Customer Call List.xlsx")
df
```

# Deleting duplicates

```python
df = df.drop_duplicates()
df
```

# Dropping needless columns

```python
df = df.drop(columns = "Not_Useful_Column")
df
```

# Normalization 

##### Removing unwanted characters like . , / _ from "Last_Name"

```python
df["Last_Name"] = df["Last_Name"].str.strip("./_")
df
```

###### removing characters like /-| between the values of Phone_Number

```python
df['Phone_Number'] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','',regex=True)
df
```

```python
# convering Phone_Numbervalues from float to string
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))

# inserting '-' between the phone numbers
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
# removing values such as nan-- / Na-- form Phone_Number
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', '')
df
```

##### Now Phone_Numbers are normalized

```python
# normalizing Paying Customer values
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')

# normalizing Do_Not_Contact values
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')

df
```

# Numerical encoding

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Y', '1')
df['Paying Customer'] = df['Paying Customer'].str.replace('N', '0')

df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Y', '1')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('N', '0')
```

```python
df
```

```python
df = df.fillna('')
df
```

```python
import matplotlib.pyplot as plt

pc = df['Paying Customer']
pc = pc[pc.isin(['0','1'])]
pc_c = pc.value_counts().sort_index()

plt.pie(pc_c, labels=pc_c.index, autopct='%1.1f%%')
plt.title('Paying Customer Distribution')
plt.show()
```

```python
dc = df['Do_Not_Contact']
dc = dc[dc.isin(['0','1'])]
dc_c = dc.value_counts().sort_index()

plt.pie(dc_c, labels=dc_c.index, autopct='%1.1f%%')
plt.title('Paying Customer Distribution')
plt.show()
```

# Visualizing and removing outlier from a dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r'C:\Extras\Programs\My Projects\customer call data cleaning\dataset\employee.csv')

plt.figure(figsize=(8, 4))
plt.boxplot(df['Salary'], vert=False)
plt.title('Boxplot of Salary')
plt.xlabel('Salary')
plt.show()

Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Salary'] < lower_bound) | (df['Salary'] > upper_bound)]
print("Outliers:\n", outliers)

df_clean = df[(df['Salary'] >= lower_bound) & (df['Salary'] <= upper_bound)]

print("\nCleaned Dataset:\n")
df_clean
```
